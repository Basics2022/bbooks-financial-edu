```python
import numpy as np

import pandas as pd
# from IPython.display import display  # import not required, maybe

import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots

import plotly.io as pio

import yfinance as yf
```

```python
#> List of tickers
tickers = [
    'SPY',        #
    'TLT',        #
    'BIL',        # 1-3 months T-Bills - SPDR
    'CSBGU3.SW',  # 1-3 Year T-Bonds (Acc) - iShares
    'GLD',        #
    # 'EURUSD=X'    #
]

# Download historical data (using 20 years for a rich dataset)
data = yf.download(tickers, start="1990-01-01", group_by="ticker") # start="2004-01-01", end="2025-12-31")
```

```python
#> Separating data in one df per asset
ticker_dict = { ticker: { "data": data[ticker].dropna() } for ticker in data.columns.levels[0] }
```

```python
"""
#> Inspection
for t in tickers:
    print(f"{t}:")
    display(ticker_dict[t]["data"].head(3))
"""
```

```python
dt_plot = 5
fig = go.Figure()
for t, t_data in ticker_dict.items():
    fig.add_trace(go.Scatter(x=t_data.index[::5], y=t_data['Close'][::5], mode='lines', name=t))

fig.update_yaxes(type="log")
fig.show()
```
